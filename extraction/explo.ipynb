```python
df = pd.read_csv("Data_Science_Jobs.csv")
df = df[df["Role_Category"] == "Analytics & BI"]
df = df[["Skill1", "Skill1", "Skill2", "Skill3", "Skill4", "Skill5", "Skll6", "Skill7", "Skill8" ]]
df
```

```python
count = 0
liste_skills = []
for _, row in df.iterrows():
    for item in row:
        count+=1
        if item not in liste_skills:
            liste_skills.append(item)

        
    
```

```python
dict_skills = {item: 0 for item in liste_skills}
```
